```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import time
Trend_Search = TrendReq()
```

```python
kw_list = ["Amazon"]
Trend_Search.build_payload(kw_list, cat=0, timeframe='today 12-m', geo='', gprop='')
time.sleep(5)
```

```python
data= Trend_Search.interest_over_time()
data.sort_values(by=['Amazon'], ascending=False, inplace=True)
data.head(10)
print(data)
```

```python
kw_list = ["Amazon"]
Trend_Search.build_payload(kw_list, cat=0, timeframe='2025-07-01 2025-08-01', geo='', gprop='')
data = Trend_Search.interest_over_time()
data = data.sort_values(by="Amazon", ascending = False)
data = data.head(10)
print(data)
```

```python
data = Trend_Search.interest_by_region()
data = data.sort_values(by="Amazon", ascending = False)
data = data.head(10)
print(data)
```

```python
data.reset_index().plot(x='geoName', y='Amazon',figsize=(10,5),kind='bar')
plt.show()
```
